````python
import csv

Date =[]
Profit_Loss = []
change_list =[]

print("Financial Analysis")
print("----------------------------")

###########################################################
with open('python-challenge/PyBank/Resources/budget_data.csv','rt')as a_file:
    x = csv.reader(a_file)
    # stores the data in list to be manipulated
    for row in x:
        Date.append(row[0])
        Profit_Loss.append((row[1]))
#     print("data is stored in Date and Profit_Loss")
    
    # crude method of removing the headers
    del Profit_Loss[0]
    del Date[0]
    ####################################################
    ## The total number of months 
    number = len(Date) #length of list 
    print("Total Months: ", number)
    ####################################################
    ## The net total amount of "Profit/Losses" over 
    ## the entire period 
    Total = 0
    for p in Profit_Loss:
        Total = Total + int(p)
    print ("Total = ", "$",Total)
    ####################################################
    ## The average of the changes in "Profit/Losses" 
    ## over the entire period
    total_changes = 0
    for i in range(0,len(Profit_Loss)-1):
        
        changes = -int(Profit_Loss[i]) + int(Profit_Loss[i+1])
        total_changes = total_changes + changes
        avg_change = total_changes/(len(Profit_Loss)-1)
        avg_change = "%.2f" % avg_change # this prints to 2 decimal places
        change_list.append(changes)
    print ("Average  Change: ",
           "$",avg_change     
          )
    ####################################################
    ## The greatest increase in profits (date and amount)
    ## over the entire period
    greatest_increase = sorted(change_list)[-1]
    greatest_increase_date = Date[change_list.index(greatest_increase)+1]
    print ("Greatest Increase in Profits: ", 
           greatest_increase_date,"$",
           greatest_increase, )
    ####################################################
    ## The greatest decrease in losses (date and amount)
    ## over the entire period
    greatest_decrease = sorted(change_list)[0]
    greatest_decrease_date = Date[change_list.index(greatest_decrease)+1]
    print ("Greatest Decrease in Profits: ",
           greatest_decrease_date,
           "$",greatest_decrease 
          )
````
